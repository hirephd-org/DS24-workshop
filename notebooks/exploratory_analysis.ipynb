```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the dataset
df = pd.read_csv('../data/happiness_report.csv')
df
```

```python

plt.figure(figsize=(12, 6))
sns.barplot(x=df.index, y='Happiness_Score', data=df, palette='viridis')
plt.xticks(rotation=45, ha='right')
plt.title('Happiness Score by Country')
plt.xlabel('Country')
plt.ylabel('Happiness Score')
plt.show()
```
